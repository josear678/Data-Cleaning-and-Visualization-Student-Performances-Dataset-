#Cleaning up data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("https://raw.githubusercontent.com/josear678/BCA_Summer_2025/refs/heads/main/student-mat.csv",sep=';')
```

```python
df2 = df
```

```python
df2['Medu'] = df2['Medu'].replace(to_replace={0:'none',1:'primary education(4th grade)',2:'5th to 9th grade',3:'secondary education',4:'higher education'})
df2['Fedu'] = df2['Fedu'].replace(to_replace={0:'none',1:'primary education(4th grade)',2:'5th to 9th grade',3:'secondary education',4:'higher education'})
df2['traveltime'] = df2['traveltime'].replace(to_replace={1:'<15min',2:'15 to 30 min',3:'30 min to 1 hour',4:'>1hour'})
df2['studytime'] = df2['studytime'].replace(to_replace={1:'<2 hours',2:'2 to 5 hours',3:'5 to 10 hours',4:'>10 hours'})
df2['famrel'] = df2['famrel'].replace(to_replace={1:'very bad', 2:'bad', 3:'neutral',4:'good',5:'excellent'})
df2['goout'] = df['goout'].replace(to_replace={1:'never',2:'barely',3:'sometimes',4:'often',5:'very often'})
df2['Dalc'] = df['Dalc'].replace(to_replace={1:'very low',2:'low',3:'mid',4:'high',5:'very high'})
df2['Walc'] = df['Walc'].replace(to_replace={1:'very low',2:'low',3:'mid',4:'high',5:'very high'})
df2['health'] = df['health'].replace(to_replace={1:'very bad',2:'bad',3:'mid',4:'good',5:'very good'})
df2['freetime'] = df['freetime'].replace(to_replace={1:'very low',2:'low',3:'mid',4:'high',5:'very high'})
```

```python
df2 = pd.get_dummies(df2, dtype=int)
df2.columns
```

#Plots

```python
#For the heatmap below I had too many columns, so I chose 15 columns that I found the most interesting
favCol = ['age','failures','absences','G1','G2','G3','address_R', 'address_U','Walc_very high','Dalc_very high','health_very bad','health_very good','famrel_excellent','famrel_very bad','activities_yes']
```

```python
sns.set()
plt.figure(figsize=(15,12))
sns.heatmap(df2[favCol].corr(), cmap='BrBG', annot=True, fmt='0.3f', center=0)
plt.title('Student Performances Heatmap')
plt.savefig('Student_Performances_Heatmap.png')
plt.show()
```

```python
sns.set()
sns.pairplot(df2[['G1','G2','G3','age','absences','failures']])

plt.title('Student Performances Pair Plot')
plt.savefig('SP_Pair_Plot.png')
plt.show()
```

```python
sns.set()
sns.histplot(data=df2, x='absences',kde=True)
plt.title('Histogram of absences')
plt.savefig('SP_Histogram_absences.png')
plt.show()
```

```python
sns.set()
sns.histplot(data=df2, x='G2',kde=True)
plt.title('Histogram of G2')
plt.savefig('SP_Histogram_G2.png')
plt.show()
```

```python
sns.set()
sns.histplot(data=df2, x='G3',kde=True)
plt.title('Histogram of G3')
plt.savefig('SP_Histogram_G3.png')
plt.show()
```

```python
sns.set()
sns.regplot(data=df2, x='G1', y='G3', line_kws={'color':'red'}, order=2)
plt.title('Regression G1 vs G3')
plt.savefig('SP_Regression_G1_vs_G3.png')
plt.show()
```

```python
sns.regplot(data=df2, x='G1', y='G2', line_kws={'color':'red'}, order=2)
plt.title('Regression G1 vs G2')
plt.savefig('SP_Regression_G1_vs_G2.png')
plt.show()
```
